```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day33-handling-mixed-variables/titanic.csv")
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passenger travelling with')
```

```python
df['number_numerical'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df['number_cateogrical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # capture numerical part
df['cabin_cat']  =df['Cabin'].str[0] # capture the first letter

df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['Ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['Ticket_num'] = pd.to_numeric(df['Ticket_num'],errors='coerce',downcast='integer')

df['Ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['Ticket_cat'] = np.where(df['Ticket_cat'].str.isdigit(),np.nan,df['Ticket_cat'])

df.head(20)
```
