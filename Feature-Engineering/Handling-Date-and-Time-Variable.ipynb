# Working With Dates

```python
# converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1. Extract Year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

2. Extract Month

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

3. Extract Days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# day of week
date['date_low'] = date['date'].dt.dayofweek
date.head()
```

```python
# day of week by name

date['date_dow_name'] = date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']),1,0)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

4. Extract Week of the year

```python
date['date-week'] = date['date'].dt.weekday
date.drop(columns=['product_id','city_id', 'orders']).head()
```

5. Extract Quater 

```python
date['Quater'] = date['date'].dt.quarter
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

6. Extract Semester

```python
date['semester'] = np.where(date['Quater'].isin([1,2]),1,2)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

7. Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()
today
```
