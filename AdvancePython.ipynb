```python
#Creating a class
#Blueprints
class Students:

  School = "Divie Sainik School"
  Address = "Lahartara Varanasi"
  school_Code = 22110825

  def __init__(self, name, marks, attendence):

    self.name = name
    self.marks = marks
    self.attendence = attendence

  def calculate_grades(self):
    if self.marks >= 90:
      return "A"
    elif self.marks >= 75:
      return "B"
    else:
      return "C"


s1 = Students("Ayush",95,90)
s2 = Students("Rohit",80,75)
s3 = Students("Satyam",28,90)

print("School Name: ",Students.School)
print("School Address: ",Students.Address)
print("School Code: ",Students.school_Code)
print()
print("Student 1 grade is : ",s1.calculate_grades())
print("Student 2 grade is : ",s2.calculate_grades())
print("Student 3 grade is : ",s3.calculate_grades())
```

```python
#Encapusulation
#to bind data and method inside the class
#noone access our internal data

class BankAccout:
  def __init__(self, name, balance):
    self.name = name
    self.__balance = balance #variable becomes private

  def deposite(self, amount):
    if amount >0:
      self.__balance += amount

  def withdraw(self, amount):
    if 0 < amount <= self.__balance:
      self.__balance -= amount
    else:
      print("Invalid withdrawl")

  def check_balance(self):
    return self.__balance


acc1 = BankAccout("Ayush", 25000)
print("1st",acc1.check_balance())
acc1.deposite(100000)
acc1.__balance = 50000 #this is a private variable
print("After deposite 5000",acc1.check_balance())
acc1.withdraw(75000)
print("After withdrwal 2000",acc1.check_balance())

#acc1.balanace = 100000
# print(acc1.name)
```

#Bank Management Mini Project

```python
class InvalidAgeError(Exception):
   pass
def check_age(age):
    if age < 18:
      raise InvalidAgeError("Age is must be 18+ or above")
    else:
      print("Access Granted")


class InvalidAmountError(Exception):
  pass
def check_amount(amount):
  if amount <= 0:
    raise InvalidAmountError("You cannot enter amount below zero")
  else:
    print("Amount Credited")


try:
  check_age(18)
except InvalidAgeError as e:
  print("Error Caught",e)
finally:
  print("Your account is credited")
  print()

try:
  check_amount(200)
except InvalidAmountError as a:
  print("Error Caught", a)
finally:
  print("Credited.....Your Amount")
```

```python
class GDRegressor:
  def __init__(self,LearningRate,epochs):
    self.m = 50.90745328
    self.b = -120
    self.lr = LearningRate
    self.epochs = epochs

  def fit(self,X,y):
    #calculate the b using GradientDescent
    for i in range(self.epochs):
      loss_slope = -2 * np.sum(y - self.m*X.ravel() - self.b)
      self.b = self.b - (self.lr * loss_slope)
      print(loss_slope,self.b)
    print(self.b)

```
